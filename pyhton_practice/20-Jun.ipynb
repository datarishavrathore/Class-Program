```python
#write a function to remove last word from string
string="this is python code in VS"
def remove_last_word(text):
    return " ".join(text.split()[:-1])

print(remove_last_word(string))
```

```python
# waf to remove word present with vowels
def remove_vowel_words(text):
    words = text.split()
    result = [word for word in words if not any(ch.lower() in "aeiou" for ch in word)]
    return " ".join(result)

print(remove_vowel_words(string))
```

```python
#waf count total space in string

def count_spaces(text):
    return text.count(" ")

print(count_spaces(string))
```

```python
# waf to remove all whitespaces
string = "This is python code in vs"

def remove_space(string):
    return string.replace(" ", "")

print(remove_space(string))
```

```python
# waf to function to print all the special syumbols
string = "This is python code@123 in version-35-address noida"

def print_special(s):
    for ch in s:
        if not ch.isalnum() and ch!="":
            print(ch)
print_special(string)
```

```python
# waf to reverse all the word of string  word by word
string = "This is python code@123 in version-35-address noida"

def reverse_words(text):
    words = text.split()
    rev = [word[::-1] for word in words]
    return " ".join(rev)

print(reverse_words(string))
```
